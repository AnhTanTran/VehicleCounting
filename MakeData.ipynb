```python
import os
import glob
import cv2
root = os.path.join("C:/Users/AT TRAN/Desktop/Make Data/image-224-Roi.jpg")
paths = []
paths.append(root)

def AnnotateImage(Image):
    # Select ROI
    showCrosshair = False
    fromCenter = False
    r = cv2.selectROI("Click or Select Bounding Box to Annotate a Vehicle", Image, fromCenter, showCrosshair)
    #Get centroid
    x=int((int(r[0])+int(r[0] + r[2]))/2) 
    y=int((int(r[1])+int(r[1] + r[3]))/2)
      
    #cv2.destroyAllWindows()
    #Return 
    if (x==0 & y==0):
        return False, 0
    point =str("")
    point = str(x) +" "+ str(y)
    print(point)
    for i in range(-2,2):
        for j in range(-2,2):
            img[y+i,x+j]=[0,0,255]
    return True,str(point)

i=0
for path in paths:    
    #go to image n-th by changing checkpoint below
    checkpoint=0
    if i < checkpoint:
        i+=1
        continue
    
    print("Open: ",path)
    #save annotation to "points"
    img = cv2.imread(path)
    points=""
    firstline=True
    flag,temp = AnnotateImage(img)
    while(flag):
        if firstline == True:
            points += temp
            firstline = False
        else:
            points += "\n" + temp
        flag,temp = AnnotateImage(img)
        continue
    print(points)
   
    #show the result
    cv2.destroyAllWindows()
    cv2.imshow("Result",img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    #write file
    file_txt = path.replace("Roi.jpg",".txt",1)
    f = open(file_txt, 'w')
    f.write(points)
    print("Saving... ",file_txt,"\nDone!")
    f.close()
    
    #Wait Key
    pause=input("Continue...")
    if pause=='':
        continue
    print("Exit")
    break
```
